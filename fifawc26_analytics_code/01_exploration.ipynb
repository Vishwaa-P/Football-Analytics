```python
def add_per90(df):
    # Avoid division by zero
    df["90s"] = df["Min"] / 90
    
    # Shooting per90
    df["Gls_per90"] = df["Gls"] / df["90s"]
    df["Ast_per90"] = df["Ast"] / df["90s"]
    df["G+A_per90"] = df["G+A"] / df["90s"]
    df["Sh_per90"] = df["Sh"] / df["90s"]
    df["SoT_per90"] = df["SoT"] / df["90s"]
    df["xG_per90"] = df["xG"] / df["90s"]
    df["xAG_per90"] = df["xAG"] / df["90s"]

    # Chance creation
    df["KP_per90"] = df["KP"] / df["90s"]
    df["CrsPA_per90"] = df["CrsPA"] / df["90s"]
    df["Carries_per90"] = df["Carries"] / df["90s"]


    # Progression
    df["PrgC_per90"] = df["PrgC"] / df["90s"]
    df["PrgP_per90"] = df["PrgP"] / df["90s"]
    df["PrgR_per90"] = df["PrgR"] / df["90s"]  # for forwards/mids
    
    # Defensive per90
    df["Tkl_per90"] = df["Tkl"] / df["90s"]
    df["Int_per90"] = df["Int"] / df["90s"]
    df["Clr_per90"] = df["Clr"] / df["90s"]
    df["Tkl+Int_per90"] = df["Tkl+Int"] / df["90s"]
    
    return df
```

##### Function to compute percentiles

```python
import numpy as np

def add_percentiles(df, metric_list):
    for col in metric_list:
        pct_col = col + "_pct"
        df[pct_col] = df[col].rank(pct=True) * 100
    return df
```

```python
metrics = [
    "Gls_per90", "Ast_per90", "G+A_per90",
    "Sh_per90", "SoT_per90",
    "xG_per90", "xAG_per90",
    "KP_per90", "CrsPA_per90",
    "PrgC_per90", "PrgP_per90", "PrgR_per90",
    "Tkl_per90", "Int_per90", 
    "Clr_per90", "Tkl+Int_per90","Carries_per90"
]
```

```python
df_fw = add_percentiles(df_fw, metrics)
df_mf = add_percentiles(df_mf, metrics)
df_fullback = add_percentiles(df_fullback, metrics)
df_centerback = add_percentiles(df_centerback, metrics)
```

```python
# ---------- LEAGUE STRENGTH ADJUSTMENT ----------

league_weights = {
    "eng Premier League": 1.00,
    "es La Liga": 0.95,
    "de Bundesliga": 0.90,
    "it Serie A": 0.90,
    "fr Ligue 1": 0.85,
    "nl Eredivisie": 0.75,
    "pt Primeira Liga": 0.75,
    "eng Championship": 0.70,
    "be Jupiler Pro League": 0.70,
    "aut Bundesliga": 0.65,
    "sui Super League": 0.65,
    "usa MLS": 0.60,
    "tr Sper Lig": 0.60,
    "other": 0.60
}

def get_league_weight(comp):
    return league_weights.get(comp, league_weights["other"])

# Apply weights
df_fw["league_weight"] = df_fw["Comp"].apply(get_league_weight)
df_mf["league_weight"] = df_mf["Comp"].apply(get_league_weight)
df_fullback["league_weight"] = df_fullback["Comp"].apply(get_league_weight)
df_centerback["league_weight"] = df_centerback["Comp"].apply(get_league_weight)
```

```python
# Adjust percentiles with league strength
def adjust_percentiles(df):
    pct_cols = [col for col in df.columns if col.endswith("_pct")]
    for col in pct_cols:
        df[col + "_adj"] = df[col] * df["league_weight"]
    return df

df_fw = adjust_percentiles(df_fw)
df_mf = adjust_percentiles(df_mf)
df_fullback = adjust_percentiles(df_fullback)
df_centerback = adjust_percentiles(df_centerback)
```

```python
def score_player(df, position):
    # FINISHING
    finish = df[[
        "Gls_per90_pct_adj", "G+A_per90_pct_adj",
        "Sh_per90_pct_adj", "SoT_per90_pct_adj",
        "xG_per90_pct_adj"
    ]].mean(axis=1)

    # CHANCE CREATION
    create = df[[
        "KP_per90_pct_adj",
        "Ast_per90_pct_adj",
        "xAG_per90_pct_adj"
    ]].mean(axis=1)

    # PROGRESSION (Carries only here, not in create)
    prog = df[[
        "Carries_per90_pct_adj",
        "PrgC_per90_pct_adj",
        "PrgP_per90_pct_adj",
        "PrgR_per90_pct_adj"
    ]].mean(axis=1)

    # DEFENSIVE CONTRIBUTION
    defend = df[[
        "Tkl_per90_pct_adj",
        "Int_per90_pct_adj",
        "Clr_per90_pct_adj",
        "Tkl+Int_per90_pct_adj"
    ]].mean(axis=1)

    # POSITION WEIGHTS
    weights = {
        "FW": (0.70, 0.15, 0.10, 0.05),
        "MF": (0.20, 0.35, 0.30, 0.15),
        "FB": (0.15, 0.30, 0.30, 0.25),
        "CB": (0.05, 0.10, 0.25, 0.60)
    }

    w_fin, w_cre, w_prog, w_def = weights[position]

    return (
        w_fin * finish +
        w_cre * create +
        w_prog * prog +
        w_def * defend
    )
```

```python
df_fw["score"] = score_player(df_fw, "FW")
df_mf["score"] = score_player(df_mf, "MF")
df_fullback["score"] = score_player(df_fullback, "FB")
df_centerback["score"] = score_player(df_centerback, "CB")
```

```python
# GK per90
def add_gk_metrics(df):
    df["Saves_per90"] = df["Saves"] / df["90s"]
    df["SoTA_per90"] = df["SoTA"] / df["90s"]
    return df

df_gk = add_gk_metrics(df_gk)

# GK percentiles
def add_gk_percentiles(df):
    df["Save_pct"] = df["Save%"].rank(pct=True) * 100
    df["PSxG_pct"] = df["PSxG+/-"].rank(pct=True) * 100
    df["CS_pct"] = df["CS%"].rank(pct=True) * 100
    df["Saves_per90_pct"] = df["Saves_per90"].rank(pct=True) * 100
    df["GA90_pct"] = (1 - df["GA90"].rank(pct=True)) * 100
    return df

df_gk = add_gk_percentiles(df_gk)

#  ADD THIS (missing earlier)
df_gk["league_weight"] = df_gk["Comp"].apply(get_league_weight)

# League-adjusted GK percentiles
gk_pct_cols = [
    "Save_pct",
    "PSxG_pct",
    "CS_pct",
    "Saves_per90_pct",
    "GA90_pct"
]

for col in gk_pct_cols:
    df_gk[col + "_adj"] = df_gk[col] * df_gk["league_weight"]
```
